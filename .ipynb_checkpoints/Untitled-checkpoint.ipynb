```python
import cv2
import imutils
import matplotlib.pyplot as plt
import numpy as np
import pytesseract
from utils.transform import rectify, four_point_transform
from utils.image import toGrayscale, denoise, thresholding, blackhat, whiteSquarify, showImage

plt.rcParams['figure.figsize'] = [8, 4]
plt.rcParams['figure.dpi'] = 100
```

```python
image = cv2.imread('./data/cni3.jpg')
image = imutils.resize(image, height=600)
showImage(image)

gray = toGrayscale(image)
blurred = denoise(gray, 3)
bh = blackhat(blurred, (13, 5))
showImage(bh, gray=True)

dst=cv2.bitwise_not(bh)
showImage(dst, gray=True)

squarified = whiteSquarify(bh)
showImage(squarified, gray=True)



cnts, hierarchy = cv2.findContours(squarified.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
# cnts = imutils.grab_contours(cnts)
cnts = sorted(cnts, key=cv2.contourArea, reverse=True)
# loop over the contours
for c in cnts:
    # compute the bounding box of the contour and use the contour to
    # compute the aspect ratio and coverage ratio of the bounding box
    # width to the width of the image
    (x, y, w, h) = cv2.boundingRect(c)
    ar = w / float(h)
    crWidth = w / float(gray.shape[1])
    # check to see if the aspect ratio and coverage width are within
    # acceptable criteria
    if ar > 5 and crWidth > 0.75:
        # pad the bounding box since we applied erosions and now need
        # to re-grow it
        pX = int((x + w) * 0.03)
        pY = int((y + h) * 0.03)
        (x, y) = (x - pX, y - pY)
        (w, h) = (w + (pX * 2), h + (pY * 2))
        # extract the ROI from the image and draw a bounding box
        # surrounding the MRZ
        roi = blurred[y:y + h, x:x + w].copy()
#         cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

        epsilon = 0.004 * cv2.arcLength(c, True)
        approx = cv2.approxPolyDP(c, epsilon, True)
        target=approx
        
        approx = rectify(target)
        print(approx)
        
        approx[0] = [x-30 for x in approx[0]]
        approx[1][0] +=30
        approx[1][1] -=30
        approx[2] = [x+30 for x in approx[2]]
        approx[3][0] -=30
        approx[3][1] +=30

        print(approx)
        
        roi = four_point_transform(image, approx)
        roi = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
        roi = cv2.GaussianBlur(roi, (1, 1), 0)
        
#         cv2.drawContours(image, [approx], 0, (0, 255, 0), 10, cv2.LINE_8, hierarchy, 0)
        
        
        break
# show the output images
plt.imshow(image)
plt.show()
plt.imshow(roi)
plt.show()
```
